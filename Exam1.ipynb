# Müşterilerden Ülkesi USA yada Brezil Olanları Bulunuz

```python
%%sql
SELECT * FROM Customers LIMIT 2;
```

```python
%%sql
SELECT CustomerName, Country FROM Customers WHERE Country = "Brazil" OR Country = "USA" ORDER BY Country;
```

```python
%%sql
SELECT CustomerName, Country FROM Customers WHERE Country in ("Brazil","USA");
```

# Önceye Ülkeye Sonra Şehre Göre Gruop By İşlemi Yapınız Sonra Bu Kırılımdan Müşteri Sayısını Bulunuz
