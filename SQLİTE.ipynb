# LIKE

- 'a%' a ile başlayan ifadeleri al

- '%a' a ile biten ifadeleri al

- '%a%' içinde a geçen ifadeleri al

- '_a%' ikinci harfi a olanları al

- 'a%%' a ile başlayan ve en az üç karakterli olanları al

- 'a%e' a ile başlayıp e ile biten ifadeleri al

```python
%%sql
SELECT * FROM Customers WHERE Country LIKE 'A%' LIMIT 10;
```

```python
%%sql 
SELECT * FROM Customers WHERE CustomerName LIKE 'A%a' ; 
```

```python
%%sql
SELECT * FROM Customers WHERE CustomerName LIKE 'F%' AND Country LIKE '%n';
```

# GROUP BY

```python
%%sql
SELECT COUNT(CustomerID) AS total_number_of_custemers, Country FROM Customers GROUP BY Country ORDER BY COUNT(CustomerName) DESC;
```

```python
%%sql
SELECT COUNT(*) AS total_num_customers, Country FROM Customers GROUP BY Country ORDER BY total_num_customers DESC;
```

```python
%%sql
SELECT AVG(Price) AS avrage_price, CategoryID FROM Products GROUP BY CategoryID;
```

```python
%%sql
SELECT AVG(Price) AS avErage_price, CategoryID FROM Products GROUP BY CategoryID ORDER BY  avErage_price DESC;
```
